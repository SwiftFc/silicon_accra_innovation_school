##### INVESTIGATING EACH COLUMN

```python
##                          BEDROOM COLUMN CLEANING PROCESS
print(f"Unique values in Bedrooms columns are: {df['Bedrooms'].nunique()} in number")
print(f"\nList the unique values\n{df['Bedrooms'].unique()}")
print(f"\nTotal Unique value count in Bedroom Column are as follow:\n {df['Bedrooms'].value_counts()}")
```

```python
###We need to take 'F' and 'A' out of the Bedroom Column and replace them with null values
df['Bedrooms'] = df['Bedrooms'].str.replace('F', '')
df['Bedrooms'] = df['Bedrooms'].str.replace('A', '')
print(f"\nList the unique values\n{df['Bedrooms'].unique()}")
```

```python
#                  HOUSE TYPE COLUMN CLEANING PROCESS
print(f"Unique values in House Type columns are: {df['House Type'].nunique()} in number")
print(f"\nList the unique values\n{df['House Type'].unique()}")
print(f"\nTotal Unique value count in House Type Column are as follow:\n {df['House Type'].value_counts()}")
```

```python
#Remove '48 m', 'hed apartment ', out of the HOUSE TYPE Column and replace them with appropriate values
df['House Type'] = df['House Type'].str.replace('48 m', '')
df['House Type'] = df['House Type'].str.replace(' furnished apartment ', 'furnish apartment')
df['House Type'] = df['House Type'].str.replace('hed apartment ', 'furnish apartment')

#df['House Type'] = df['House Type'].str.replace('ent ', '')


print(f"\nList the unique values\n{df['House Type'].unique()}")
```

```python
print(f"\nTotal Unique value count in House Type Column are as follow:\n {df['House Type'].value_counts()}")
```

```python
#                            LOCATION COLUMN CLEANING PROCESS
print(f"Unique values in Location columns are: {df['Location'].nunique()} in number")
print(f"\nList the unique values\n{df['Location'].unique()}")
```

```python
print(f"\nTotal Unique value count in Location Column are as follow:\n {df['Location'].value_counts().head(60)}")
```

```python
df['Location/ Suburb'] = df['Location/ Suburb'].str.replace('1 1 48 m', 'NaN')
```

```python
#                       PRICE COLUMN CLEANING PROCESS
df['Price'].nunique()
```

```python
df['Price'] = df['Price'].str.replace('1 1 1 48 m', 'NaN')
```

```python
df['Price'] = df['Price'].str.replace('NaN', "")
df['Price'] = df['Price'].str.replace('[Price disclosed on request]', "")
```

```python
for x in df.index:
    if df.loc[x, 'Price'] == "":
        df.drop(x, inplace=True)
df
```

```python
for x in df.index:
    if df.loc[x, 'Bedrooms'] == "":
        df.drop(x, inplace=True)
df
```

```python
df
```

```python
df['Price'] = pd.to_numeric(df['Price'], errors='coerce',downcast='integer')
```

```python
df = df.reset_index(drop=True)
df
```

```python
IQR = df['Price'].quantile(0.75) - df['Price'].quantile(0.25)
print("The inter quantile range for the price column is", IQR)
```

```python
low_outlier = df['Price'].quantile(0.25) - 1.5 *IQR
print("The low outlier for price is", low_outlier)
```

```python
high_outlier = df['Price'].quantile(0.75) + 1.5 *IQR
print("The high outlier for the price column is", high_outlier)
```
