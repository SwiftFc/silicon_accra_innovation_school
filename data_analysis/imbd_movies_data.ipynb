### 20. Classify Movies Based on Ratings [Excellent, Good and Average]

```python

def rating(rating):   
    if rating>=7.0:
        return 'Excellent'
    elif rating>=6.0:
        return 'Good'
    else:
        return 'Average'

data['Score'] = data['Rating'].apply(rating)
```

```python
data[['Title', 'Rating', 'Score']].sample(10). \
set_index("Title")
```
