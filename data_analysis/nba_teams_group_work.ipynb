### import data

```python
import pandas as pd
import numpy as np



golden_state = pd.read_html("https://www.espn.com/nba/team/stats/_/name/gs/golden-state-warriors")
miami_heat = pd.read_html("https://www.espn.com/nba/team/stats/_/name/mia/miami-heat")
boston = pd.read_html("https://www.espn.com/nba/team/stats/_/name/bos/boston-celtics")
```

### slice data

```python
golden_state = pd.concat((golden_state[2],golden_state[1]), axis = 1)
miami_heat = pd.concat((miami_heat[2], miami_heat[1]), axis = 1)
boston = pd.concat((boston[2], boston[1]), axis = 1)

golden_state = golden_state[:15]
miami_heat = miami_heat[:17]
boston = boston[:13]
```

### create column for Team names

```python
golden_state['Team'] = 'Golden State Warriors'
miami_heat['Team'] = 'Miami Heat'
boston['Team'] = 'Boston'
```

### merge tables

```python
df = pd.concat((golden_state, miami_heat, boston), axis = 0, ignore_index = True)
df
```

### Team with the MVP

```python
df = df.replace([np.inf],0)
```

```python
col = df.iloc[0:,1:14]
```

```python
df['Player_Stats'] = col.sum(axis =1)/100
```
