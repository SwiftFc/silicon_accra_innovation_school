```python
import pandas as pd
```

```python
golden_state = pd.read_html("https://www.espn.com/nba/team/stats/_/name/gs/season/2023/seasontype/2")
golden_state_data = golden_state[0]
golden_state_data['Team Name'] = "Golden State Warriors"
golden_state_data
```

```python
golden_state_data = golden_state[1]
golden_state_data = pd.concat([golden_state[0],golden_state[1]], axis = 1)
g = golden_state_data.drop(18)
g
```

### Miami Heat Statitics

```python
miami_heat = pd.read_html("https://www.espn.com/nba/team/stats/_/name/mia/season/2023/seasontype/2")
miami_heat_data = miami_heat[0]
miami_heat_data['Team Name'] = 'Miami Heat'
miami_heat_data
```

```python
miami_heat_data = miami_heat[1]
miami_heat_data = pd.concat([miami_heat[0], miami_heat[1]], axis = 1)
y = miami_heat_data.drop(20)
y
```

### Boston Celtics Statistics

```python
boston_celtics = pd.read_html("https://www.espn.com/nba/team/stats/_/name/bos/season/2023/seasontype/2")
boston_celtics_data = boston_celtics[0]
boston_celtics_data['Team Name'] = 'Boston Celtics'
boston_celtics_data
```

```python
boston_celtics_data = boston_celtics[1]
boston_celtics_data = pd.concat([boston_celtics[0], boston_celtics[1]], axis = 1)
x= boston_celtics_data.drop(19)
x
```

### Merging all three Nba Teams

```python
nba_teams = pd.concat([g, y, x], axis = 0, ignore_index = True)
nba_teams
```

### Q3. Which team has the most Mvp Players.. ie which team has the best performing players

```python
team_points = nba_teams.groupby('Team Name')['PTS'].sum().reset_index()
team_points
```

```python
mvp = team_points[team_points['PTS'] ==team_points['PTS'].max()]
mvp
```

```python
print(f"{mvp}")
```

### Finding the most valuable player amongst the three teams

```python
golden_state_mvps = golden_state_data.head(6).groupby('Name')['PTS'].max().reset_index()
golden_state_mvps
```

```python
miami_heat_mvps = miami_heat_data.head(6).groupby('Name')['PTS'].max().reset_index()
miami_heat_mvps
```

```python
boston_celtics_mvps = boston_celtics_data.head(6).groupby('Name')['PTS'].max().reset_index()
boston_celtics_mvps
```

### Merging the top 5 most valuable players in each of the three teams based on their average points 

```python
all_team_mvps = pd.concat([golden_state_mvps, miami_heat_mvps, boston_celtics_mvps], axis = 0, ignore_index = True)
all_team_mvps
```

### Finding the overall most valuable player, based on thier average points

```python

mvp = all_team_mvps.loc[all_team_mvps['PTS'].idxmax(), 'Name']
```

```python
print(f"most valuable player amongst all three teams is {mvp}")
```

### Q4. Which 5 names occur frequently in each team and plot a bar graph

```python
import matplotlib.pyplot as plt
```

```python
team_name_counts = nba_teams['Name'].groupby(nba_teams['Team Name']).value_counts().reset_index(name='Counts')
```

```python
top_5_names_per_team = team_name_counts.groupby('Team Name').head(5)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
top_5_names_per_team.pivot(index='Team Name', columns='Name', values='Counts').plot(kind='bar', stacked=True, ax=ax)
plt.title("Top 5 Most Frequently Occurring Player Names in Each Team")
plt.xlabel("Team Name")
plt.ylabel("Counts")
plt.legend(title="Player Name", loc='upper right', bbox_to_anchor=(1.3, 1))
plt.show()
```
