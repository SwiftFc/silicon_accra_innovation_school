### Q3 What is the total Revenue made on Weekend vs Weekday

```python


# Calculate total revenue for each order
df['total_revenue'] = df['cost_of_the_order']

# Group the data by 'day_of_the_week' and sum the total revenue
revenue_by_day = df.groupby('day_of_the_week')['total_revenue'].sum()

# Print the result
print(revenue_by_day)
```

### Q4.  The restarant wants to give a free coupon to their top customer. Identify a criteria from the dataset which they could use to find their top customer to give a coupon

```python
df


# Group the data by 'customer_id' and sum the total spending
best_customer = df.groupby('customer_id')['cost_of_the_order'].sum()

# Identify the customer with the highest total spending
top_customer = best_customer.idxmax()


# Print the top customer's ID
print(f"The top customer with the highest total spending is customer ID: {top_customer}")


# Optionally, you can reward the top customer with a coupon or take other actions.
```
